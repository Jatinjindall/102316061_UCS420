##### Part I: Randomized Sales Data Generation
##### Step 1: Initialize Random Seed
Use your roll number as the random seed to ensure unique data.

```python
import numpy as np

# Replace with your roll number
roll_number = 12345
np.random.seed(roll_number)
```

##### Step 2: Generate Sales Data
Create a 12x4 NumPy array where each column represents sales data for a product category.

```python
sales_data = np.random.randint(1000, 5001, size=(12, 4))
print("Generated Sales Data:")
print(sales_data)
```

##### Step 3: Convert Data to DataFrame
Convert the generated NumPy array into a Pandas DataFrame with appropriate labels.

```python
import pandas as pd

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
categories = ["Electronics", "Clothing", "Home & Kitchen", "Sports"]
sales_df = pd.DataFrame(sales_data, columns=categories, index=months)
print(sales_df)
```

#### Part II: Data Manipulation and Analysis
##### Step 1: Display the First 5 Rows and Summary Statistics.

```python
print(sales_df.head())
print(sales_df.describe())
```

##### Step 2: Calculate Total Sales Per Category and Month.

```python
sales_df["Total Sales"] = sales_df.sum(axis=1)
print(sales_df["Total Sales"])
```

##### Step 3: Calculate Average Sales Growth Between Consecutive Months.

```python
sales_growth = sales_df.diff().mean()
print(sales_growth)
```

##### Step 4: Add Growth Rate Column.

```python
sales_df["Growth Rate"] = sales_df["Total Sales"].pct_change() * 100
print(sales_df)
```

##### Step 5: Apply a Randomized Discount Based on Roll Number.

```python
if roll_number % 2 == 0:
sales_df["Electronics"] *= 0.9  # 10% discount
else:
sales_df["Clothing"] *= 0.85  # 15% discount
print(sales_df)
```

#### Part III: Data Visualization
##### Step 1: Monthly Sales Trends (Line Plot)

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
for category in categories:
plt.plot(sales_df.index, sales_df[category], marker='o', label=category)
plt.title("Monthly Sales Trends")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()
plt.grid()
plt.show()
```

##### Step 2: Sales Distribution (Box Plot)

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=sales_df[categories])
plt.title("Sales Distribution by Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.show()
```

```python
percentiles = np.percentile(array, [25, 50, 75], axis=None)
print(percentiles)
```

##### Step 3: Compute Mean, Median, and Standard Deviation.

```python
mean_flat = np.mean(array)
median_flat = np.median(array)
std_flat = np.std(array)
print(mean_flat, median_flat, std_flat)
```

#### Question 3: For the array: a = np.array([-1.8, -1.6, -0.5, 0.5,1.6, 1.8, 3.0]).
 i. Find floor, ceiling and truncated value, rounded values.

```python
arr = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])
print("Floor:", np.floor(arr))
print("Ceiling:", np.ceil(arr))
print("Truncated:", np.trunc(arr))
print("Rounded:", np.round(arr))
```
