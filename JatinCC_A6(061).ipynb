Q1. Ask the user to enter a value (e.g., M) for a mathematical function.

Generate x values from -10 to 10 using np.linspace().

Compute y values for:

y= M⋅x^2

y=M⋅sin(x)

Plot both functions in a single figure: Use different colors and line styles. Add legend, grid, and title.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

M = float(input("Enter a value for M: "))


x = np.linspace(-10, 10, 100)


y1 = M * x**2
y2 = M * np.sin(x)


plt.figure(figsize=(10, 5))
plt.plot(x, y1, 'b--', label='y = M * x^2')
plt.plot(x, y2, 'r-', label='y = M * sin(x)')
plt.legend()
plt.xlabel('x values')
plt.ylabel('y values')
plt.title('Mathematical Functions Plot')
plt.grid(True)
plt.show()

```

Q.2 Create a dataset of five subjects and scores.

---
Convert it into a Pandas DataFrame.
Plot the scores using a Seaborn bar plot with:

Different colors for each bar.
Annotations on top of each bar.
Title, axis labels, and grid.


```python

subjects = ['Math', 'Physics', 'Chemistry', 'Computer Science', 'English']
scores = [95, 88, 95, 88, 94]


df = pd.DataFrame({'Subjects': subjects, 'Scores': scores})

# Plot using Seaborn
plt.figure(figsize=(8, 5))
sns.barplot(x='Subjects', y='Scores', data=df, palette='viridis')

# Add annotations
for index, value in enumerate(scores):
    plt.text(index, value + 1, str(value), ha='center', fontsize=12)

# Labels and Title
plt.xlabel("Subjects")
plt.ylabel("Scores")
plt.title("Scores in Different Subjects")
plt.grid(axis='y')

plt.show()
```

Q3. Write a Python script to select your roll number as seed for NumPy and
then generate a dataset of 50 values using np.random.randn().

Create a 2x2 subplot layout displaying:
Line plot showing cumulative sum.
Scatter plot with random noise.
Customize with titles, axis labels, and grids.

```python
import numpy as np
import matplotlib.pyplot as plt

roll_number = 102316061
np.random.seed(roll_number)

data = np.random.randn(50)

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

axes[0, 0].plot(np.cumsum(data), marker='o', linestyle='-', color='b')
axes[0, 0].set_title("Cumulative Sum Plot")
axes[0, 0].set_xlabel("Index")
axes[0, 0].set_ylabel("Cumulative Sum")
axes[0, 0].grid(True, linestyle='--', alpha=0.6)

axes[0, 1].scatter(range(50), data, color='r', alpha=0.7)
axes[0, 1].set_title("Scatter Plot with Random Noise")
axes[0, 1].set_xlabel("Index")
axes[0, 1].set_ylabel("Value")
axes[0, 1].grid(True, linestyle='--', alpha=0.6)

fig.delaxes(axes[1, 0])
fig.delaxes(axes[1, 1])

plt.tight_layout()
plt.show()
```

Q.4 Download Data-set from the below link
https://github.com/AnjulaMehto/MCA/blob/main/company_sales_data.csv
Apply ‘seaborn’ library to do the following.
1. Read Total profit of all months and show it using a line plot.
2. Read all product sales data and show it using a multiline plot.
3. Plot bar chart for all the features/attributes.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

url = 'https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv'
df = pd.read_csv(url)

df['month_number'] = df['month_number'].astype(str)

# 1. Line plot for Total Profit per month
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='month_number', y='total_profit', marker='o', color='b')
plt.title('Total Profit per Month')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.grid(True)
plt.show()

# 2. Multiline plot for all product sales data
plt.figure(figsize=(12, 8))
product_columns = df.columns[1:-1]
for product in product_columns:
    sns.lineplot(data=df, x='month_number', y=product, marker='o', label=product)
plt.title('Monthly Sales Data for All Products')
plt.xlabel('Month Number')
plt.ylabel('Sales Units')
plt.legend(title='Products')
plt.grid(True)
plt.show()

# 3. Bar charts for all features/attributes
df_melted = df.melt(id_vars='month_number', var_name='Feature', value_name='Value')

plt.figure(figsize=(14, 10))
sns.barplot(data=df_melted, x='month_number', y='Value', hue='Feature')
plt.title('Monthly Values for All Features')
plt.xlabel('Month Number')
plt.ylabel('Value')
plt.legend(title='Features', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```
